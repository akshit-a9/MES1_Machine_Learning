```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of sigmoid function
def sigmoid_derivative(x):
    sx = sigmoid(x)
    return sx * (1 - sx)

# Test the function
x = np.array([])
print("Sigmoid values:", sigmoid(x))
print("Sigmoid derivative values:", sigmoid_derivative(x))
```
